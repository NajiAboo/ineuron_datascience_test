```python
import pandas as pd
```

```python
df_json = pd.read_json("https://data.nasa.gov/resource/y77d-th95.json")
```

```python
df_json
```

```python
df_mid = pd.json_normalize(df_json.to_dict("records"))
```

```python
df_final =df_mid[["name", "id","nametype","mass","year","recclass", "reclat","reclong","geolocation.coordinates"]].rename(columns={"geolocation.coordinates":"point coordinates"})
```

```python
df_final
```

```python
df_final.to_csv("final.csv")
```
